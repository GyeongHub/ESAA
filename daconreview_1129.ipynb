# 코드 흐름

1. Dataset 구성
    - 주어진 데이터로 학습 후 2주 후의 12주 데이터를 예측해야하는 상황임으로 위와 같이 구성
    - 24주 데이터, 2주 후 12주 데이터
    - 최근 35년 데이터 (이전 데이터에는 유실된 데이터도 존재함)

image.png

2. 모델 구현
- 시간별 이미지 데이터 > 이미지별로 서로 영향을 준다
- Seq2Seq + ConvLSTM

3. 모델 학습
- 학습 정의
- 최적의 하이퍼 파라미터 정의
- 훈련 정의
- 체크 포인트 정의
- 학습 시작

4. 결과

- 치별점, 배울점
    - 작년과 올해 데이터의 12주차를 비교한 점
    - 예측과 결과 생성 함수를 정의하여 사용한 점
